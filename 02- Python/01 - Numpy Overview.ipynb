## How to Increase brightness in image using numpy
### Reading Image 

```python
from PIL import Image
img = Image.open("./data/image.jpg")
img
```

```python
img = np.asarray(img)
img
```

```python
plt.imshow(img)
```

## Replace values greater than 127 with 255 (Increase Brightness in Image)

```python
# Red Green Blue
red = img[:,:,0].copy()
green = img[:,:,1].copy()
blue = img[:,:,2].copy()
```

```python
red[red>127] = 255
green[green>127] = 255
blue[blue>127] = 255
```

```python
img_new = img.copy()
```

```python
img_new[:,:,0] = red
img_new[:,:,1] = green
img_new[:,:,2] = blue
```

```python
plt.imshow(img_new)
plt.show()
```
